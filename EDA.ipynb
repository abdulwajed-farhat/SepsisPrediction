```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/users/farhat/Documents/Project/ProcessedData/fullData.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
hospA = data[data['Patient_Id'].str.startswith('p0')]
hospB = data[data['Patient_Id'].str.startswith('p1')]
```

```python
100*(hospB['SepsisLabel'].value_counts()[1]/hospB['SepsisLabel'].count())
```

```python
# Add a column to identify hospital
hospA['Hospital'] = 'A'
hospB['Hospital'] = 'B'

# Combine datasets
combined = pd.concat([hospA, hospB])
combined.head(5)
```

```python
hospB.groupby('Patient_Id')['SepsisLabel'].max().value_counts()
```

```python
# Plot using hue to compare
sns.countplot(x='SepsisLabel', hue='Hospital', data=combined, edgecolor='black')

plt.xlabel('SepsisLabel', fontsize=12, fontname='Times New Roman')
plt.ylabel('Count', fontsize=12, fontname='Times New Roman')
plt.title('Distribution of SepsisLabel in Hospitals A and B', fontsize=14, fontname='Times New Roman')

plt.show()

plt.savefig('Distribution of SepsisLabel in Hospitals A and B.png')
```

```python
nans = pd.DataFrame(columns=['column', 'count'])
nans['column'] = combined.columns[1:41]
nans['count'] = list(combined.isna().sum()[1:41][0:40])
```

```python
import os
source_folder = '/Users/farhat/Documents/Project/RawData/TrainingSet_A'
hostA_raw = pd.DataFrame()
for filename in os.listdir(source_folder):
    if filename.endswith('.psv'):
        file_path = os.path.join(source_folder, filename)
        df = pd.read_csv(file_path, sep='|')
        hostA_raw = pd.concat([hostA_raw, df])
```

```python
source_folder = '/Users/farhat/Documents/Project/RawData/TrainingSet_B'
hostB_raw = pd.DataFrame()
for filename in os.listdir(source_folder):
    if filename.endswith('.psv'):
        file_path = os.path.join(source_folder, filename)
        df = pd.read_csv(file_path, sep='|')
        hostB_raw = pd.concat([hostB_raw, df])
```

```python
combined_raw = pd.concat([hostA_raw,hostB_raw])
combined_raw.head(2)
```

```python
nans_raw = pd.DataFrame(columns=['column', 'count'])
nans_raw['column'] = combined_raw.columns[:40]
nans_raw['count'] = list(combined_raw.isna().sum()[:40][0:40])
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.figure(figsize=(10, 6))

# Barplot for raw data
sns.barplot(x='column', y='count', data=nans_raw, edgecolor='black', color='skyblue', label='Raw Data')

# Barplot for processed data
sns.barplot(x='column', y='count', data=nans, edgecolor='black', color='salmon', alpha=0.7, label='Processed Data')

# Formatting
plt.xlabel('Features', fontsize=12, fontname='Times New Roman')
plt.ylabel('NaN Count', fontsize=12, fontname='Times New Roman')
plt.title('Comparison of NaN Values in Raw vs Processed Data', fontsize=14, fontname='Times New Roman')
plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python

data = combined[combined['SepsisLabel'] == 1].iloc[:, 1:-1]
data = data.iloc[:,:34]
col = 10
row = int(np.ceil(data.shape[1] / col))

fig, axes = plt.subplots(row, col, figsize=(20, 5*row))  # adjust size for readability
axes = axes.flatten()  # flatten in case of multiple rows

for i in range(data.shape[1]):
    data_col = data.iloc[:, i].dropna()
    data_col = pd.DataFrame(data_col)
    axes[i].boxplot(data_col)
    axes[i].set_title(data.columns[i], fontsize=10)

# Turn off unused subplots
for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()

```
